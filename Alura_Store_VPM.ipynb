```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

##**1. Ingreso total por cada tienda_Análisis de facturación**

```python
# Ingreso total de cada tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
# Mostrar los resultados
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)
```

##**3. Calificación promedio de la tienda**

```python
print(tienda.columns)
```

```python
# Calcular calificación promedio de cada tienda
prom_t1 = tienda['Calificación'].mean()
prom_t2 = tienda2['Calificación'].mean()
prom_t3 = tienda3['Calificación'].mean()
prom_t4 = tienda4['Calificación'].mean()

# Mostrar resultados redondeados a 2 decimales
print("Calificación promedio Tienda 1:", round(prom_t1, 2))
print("Calificación promedio Tienda 2:", round(prom_t2, 2))
print("Calificación promedio Tienda 3:", round(prom_t3, 2))
print("Calificación promedio Tienda 4:", round(prom_t4, 2))
```

##**Productos mas y menos vendidos**

```python
# Tienda 1
print("Tienda 1")
print("Más vendido:", tienda['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda['Producto'].value_counts().idxmin())
print()

# Tienda 2
print("Tienda 2")
print("Más vendido:", tienda2['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda2['Producto'].value_counts().idxmin())
print()

# Tienda 3
print("Tienda 3")
print("Más vendido:", tienda3['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda3['Producto'].value_counts().idxmin())
print()

# Tienda 4
print("Tienda 4")
print("Más vendido:", tienda4['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda4['Producto'].value_counts().idxmin())
```

##**5. Envío promedio por tienda**

```python
# Promedio de costo de envío en cada tienda
prom_envio_t1 = tienda['Costo de envío'].mean()
prom_envio_t2 = tienda2['Costo de envío'].mean()
prom_envio_t3 = tienda3['Costo de envío'].mean()
prom_envio_t4 = tienda4['Costo de envío'].mean()

# Mostrar resultados redondeados a 2 decimales
print("Costo de envío promedio Tienda 1:", round(prom_envio_t1, 2))
print("Costo de envío promedio Tienda 2:", round(prom_envio_t2, 2))
print("Costo de envío promedio Tienda 3:", round(prom_envio_t3, 2))
print("Costo de envío promedio Tienda 4:", round(prom_envio_t4, 2))
```
